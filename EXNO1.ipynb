```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
age=[1,3,28,27,25,92,30,39,40,50,26,24,29,94]
af=pd.DataFrame(age)
af
```

```python
boxplot=sns.boxplot(data=af)
```

```python
q1=af.quantile(0.25)
q2=af.quantile(0.5)
q3=af.quantile(0.75)
IQR= q3 -q1
IQR
```

```python
Q1=np.percentile(af,25)
Q2=np.percentile(af,50)
Q3=np.percentile(af,75)
IQR=Q3-Q1
IQR
```

```python
lowerbound= Q1-1.5*IQR
upperbound= Q3+1.5*IQR
```

```python
lowerbound
```

```python
upperbound
```

```python
outliers = [x for x in age if x < lowerbound or x > upperbound]
```

```python
print("Q1: ",Q1)
print("Q2: ",Q2)
print("Q3: ",Q3)
print("IQR: ",IQR)
print("Lower Bound: ",lowerbound)
print("Upper Bound: ",upperbound)
print("Outliers: ",outliers)
```

```python
outliers=af[((af>=lowerbound)&(af<=upperbound))]
outliers
```

```python
removed_outliers=outliers.dropna()
removed_outliers
```

```python
refined_boxplot=sns.boxplot(data=removed_outliers)
```
